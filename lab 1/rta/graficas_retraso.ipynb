# Análisis de temperaturas y señales de calentamiento

Este notebook lee `data_Q1_Q2.txt`, `data_Q1.txt` y `data_Q2.txt` para generar:
1. Temperaturas `Temp1` y `Temp2` vs tiempo.
2. Señales `Cal1` y `Cal2` vs tiempo.
3. Visualizaciones de retraso con doble eje (Cal vs Temp).
4. Gráficas separadas para escenarios Q1 y Q2.

Con estas dos gráficas se puede explicar el **retraso** entre el cambio en la acción de control (`Cal`) y la respuesta de temperatura.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar archivos correctos
#df_q1_q2 = cargar_archivo('/content/data_Q1_Q2.txt')
df_q1_q2 = pd.read_csv('data_Q1_Q2.txt')
df_q1 = pd.read_csv('data_Q1.txt')
df_q2 = pd.read_csv('data_Q2.txt')

# Renombrar columnas para facilitar uso
df_q1_q2.columns = ['Tiempo', 'Cal1', 'Cal2', 'Temp1', 'Temp2', 'Set1', 'Set2']
df_q2.columns = ['Tiempo', 'Cal1', 'Cal2', 'Temp1', 'Temp2', 'Set1', 'Set2']
df_q1.columns = ['Tiempo', 'Cal1', 'Cal2', 'Temp1', 'Temp2', 'Set1', 'Set2']

print('Datos cargados correctamente:')
print(f"- data_Q1_Q2.txt:   {df_q1_q2.shape}")
print(f"- data_Q1.txt: {df_q1.shape}")
print(f"- data_Q2.txt: {df_q2.shape}")

df_q1_q2.head()
```

```python
# Vista rápida de datos individuales
print('Primeras filas de Q1:')
display(df_q1.head())

print('Primeras filas de Q2:')
display(df_q2.head())
```

```python
# Gráfica 1: Temperaturas en el tiempo (Q1_Q2)
plt.figure(figsize=(10,5))
plt.plot(df_q1_q2['Tiempo'], df_q1_q2['Temp1'], label='Temp1 (°C)', linewidth=2)
plt.plot(df_q1_q2['Tiempo'], df_q1_q2['Temp2'], label='Temp2 (°C)', linewidth=2)
plt.xlabel('Tiempo (s)')
plt.ylabel('Temperatura (°C)')
plt.title('Temperaturas vs Tiempo (Q1_Q2)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Gráfica 2: Calentadores en el tiempo (Q1_Q2)
plt.figure(figsize=(10,5))
plt.step(df_q1_q2['Tiempo'], df_q1_q2['Cal1'], where='post', label='Cal1 (%)', linewidth=2)
plt.step(df_q1_q2['Tiempo'], df_q1_q2['Cal2'], where='post', label='Cal2 (%)', linewidth=2)
plt.xlabel('Tiempo (s)')
plt.ylabel('Calentador (%)')
plt.title('Cal1 y Cal2 vs Tiempo (Q1_Q2)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# Gráfica 3: Cal y Temp en una sola figura con doble eje (para ver claramente el retraso)
fig, ax1 = plt.subplots(figsize=(12, 6))

# Eje izquierdo: Temperaturas
ax1.plot(df_q1_q2['Tiempo'], df_q1_q2['Temp1'], 'b-', linewidth=2, label='Temp1')
ax1.plot(df_q1_q2['Tiempo'], df_q1_q2['Temp2'], 'r-', linewidth=2, label='Temp2')
ax1.set_xlabel('Tiempo (s)', fontsize=11)
ax1.set_ylabel('Temperatura (°C)', color='k', fontsize=11)
ax1.tick_params(axis='y')
ax1.grid(True, alpha=0.3)

# Eje derecho: Calentadores
ax2 = ax1.twinx()
ax2.step(df_q1_q2['Tiempo'], df_q1_q2['Cal1'], where='post', color='g', linewidth=2, label='Cal1', linestyle='--')
ax2.step(df_q1_q2['Tiempo'], df_q1_q2['Cal2'], where='post', color='orange', linewidth=2, label='Cal2', linestyle='--')
ax2.set_ylabel('Calentador (%)', color='k', fontsize=11)
ax2.tick_params(axis='y')

plt.title('Visualización del Retraso: Cal vs Temp', fontsize=12, fontweight='bold')

# Leyenda combinada
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

fig.tight_layout()
plt.show()

print("\n✓ Las gráficas permiten ver claramente:")
print("  - El cambio en Cal1 y Cal2 (líneas punteadas) ocurre alrededor de t=10s")
print("  - Las temperaturas comienzan a subir (líneas sólidas) después de ese momento")
print("  - Este desfase es el RETRASO del sistema termodinámico.")
```
