# Análisis de temperaturas y señales de calentamiento

Este notebook lee `data.txt` y genera:
1. Temperaturas `Temp1` y `Temp2` vs tiempo.
2. Señales `Cal1` y `Cal2` vs tiempo.

Con estas dos gráficas se puede explicar el **retraso** entre el cambio en la acción de control (`Cal`) y la respuesta de temperatura.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer archivo (este notebook está en la misma carpeta que data.txt)
df = pd.read_csv('data.txt')

# Renombrar columnas para facilitar uso
df.columns = ['Tiempo', 'Cal1', 'Cal2', 'Temp1', 'Temp2', 'Set1', 'Set2']

df.head()
```

```python
# Gráfica 1: Temperaturas en el tiempo
plt.figure(figsize=(10,5))
plt.plot(df['Tiempo'], df['Temp1'], label='Temp1 (°C)', linewidth=2)
plt.plot(df['Tiempo'], df['Temp2'], label='Temp2 (°C)', linewidth=2)
plt.xlabel('Tiempo (s)')
plt.ylabel('Temperatura (°C)')
plt.title('Temperaturas vs Tiempo')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Gráfica 2: Calentadores en el tiempo
plt.figure(figsize=(10,5))
plt.step(df['Tiempo'], df['Cal1'], where='post', label='Cal1 (%)', linewidth=2)
plt.step(df['Tiempo'], df['Cal2'], where='post', label='Cal2 (%)', linewidth=2)
plt.xlabel('Tiempo (s)')
plt.ylabel('Calentador (%)')
plt.title('Cal1 y Cal2 vs Tiempo')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# Gráfica 3: Cal y Temp en una sola figura con doble eje (para ver claramente el retraso)
fig, ax1 = plt.subplots(figsize=(12, 6))

# Eje izquierdo: Temperaturas
ax1.plot(df['Tiempo'], df['Temp1'], 'b-', linewidth=2, label='Temp1')
ax1.plot(df['Tiempo'], df['Temp2'], 'r-', linewidth=2, label='Temp2')
ax1.set_xlabel('Tiempo (s)', fontsize=11)
ax1.set_ylabel('Temperatura (°C)', color='k', fontsize=11)
ax1.tick_params(axis='y')
ax1.grid(True, alpha=0.3)

# Eje derecho: Calentadores
ax2 = ax1.twinx()
ax2.step(df['Tiempo'], df['Cal1'], where='post', color='g', linewidth=2, label='Cal1', linestyle='--')
ax2.step(df['Tiempo'], df['Cal2'], where='post', color='orange', linewidth=2, label='Cal2', linestyle='--')
ax2.set_ylabel('Calentador (%)', color='k', fontsize=11)
ax2.tick_params(axis='y')

plt.title('Visualización del Retraso: Cal vs Temp', fontsize=12, fontweight='bold')

# Leyenda combinada
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

fig.tight_layout()
plt.show()

print("\n✓ Las gráficas permiten ver claramente:")
print("  - El cambio en Cal1 y Cal2 (líneas punteadas) ocurre alrededor de t=10s")
print("  - Las temperaturas comienzan a subir (líneas sólidas) después de ese momento")
print("  - Este desfase es el RETRASO del sistema termodinámico.")
```
