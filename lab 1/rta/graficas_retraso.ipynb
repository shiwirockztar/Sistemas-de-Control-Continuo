# Análisis de temperaturas y señales de calentamiento

Este notebook lee `data_Q1_Q2.txt`, `data_Q1.txt` y `data_Q2.txt` para generar:
1. Temperaturas `Temp1` y `Temp2` vs tiempo.
2. Señales `Cal1` y `Cal2` vs tiempo.
3. Visualizaciones de retraso con doble eje (Cal vs Temp).
4. Gráficas separadas para escenarios Q1 y Q2.

Con estas dos gráficas se puede explicar el **retraso** entre el cambio en la acción de control (`Cal`) y la respuesta de temperatura.

## Estimación del modelo FOPDT

Para cada ensayo se estima un modelo de la forma:

$$G(s)=\frac{K}{\tau s + 1} e^{-\theta s}$$

- `K`: ganancia estática.
- `τ`: constante de tiempo.
- `θ`: tiempo muerto.

Se calculan parámetros por tres vías:
1. Ziegler-Nichols (tangente en máxima pendiente).
2. Smith (puntos 28.3% y 63.2% de la respuesta).
3. Optimización numérica para minimizar el error cuadrático.

```python
def moving_average(y, window=7):
    """Suaviza la señal para calcular derivadas de forma robusta."""
    if window <= 1:
        return y.copy()
    w = np.ones(window) / window
    y_pad = np.pad(y, (window // 2, window - 1 - window // 2), mode='edge')
    return np.convolve(y_pad, w, mode='valid')


def first_crossing_time(t, y, target, start_idx, direction=1):
    """Retorna el primer tiempo donde y cruza target, con interpolación lineal."""
    cond = (lambda v: v >= target) if direction >= 0 else (lambda v: v <= target)
    for i in range(max(1, start_idx), len(y)):
        y0, y1 = y[i - 1], y[i]
        if cond(y1):
            if y1 == y0:
                return t[i]
            frac = (target - y0) / (y1 - y0)
            frac = np.clip(frac, 0.0, 1.0)
            return t[i - 1] + frac * (t[i] - t[i - 1])
    return np.nan


def detect_single_step(t, u):
    du = np.diff(u)
    idx_step = int(np.argmax(np.abs(du)) + 1)
    n0 = max(3, int(0.05 * len(u)))
    nf = max(3, int(0.10 * len(u)))
    u0 = float(np.mean(u[:n0]))
    uf = float(np.mean(u[-nf:]))
    delta_u = uf - u0
    t0 = float(t[idx_step])
    return idx_step, t0, u0, uf, delta_u


def fopdt_response(t, K, tau, theta, t0, delta_u, y0):
    tau = max(float(tau), 1e-6)
    theta = max(float(theta), 0.0)
    tt = t - (t0 + theta)
    dyn = np.where(tt > 0, 1.0 - np.exp(-tt / tau), 0.0)
    return y0 + K * delta_u * dyn


def estimate_fopdt_methods(df, u_col, y_col, tag='Sistema'):
    t = df['Tiempo'].to_numpy(dtype=float)
    u = df[u_col].to_numpy(dtype=float)
    y = df[y_col].to_numpy(dtype=float)

    idx_step, t0, u0, uf, delta_u = detect_single_step(t, u)
    if abs(delta_u) < 1e-9:
        raise ValueError(f'No se detectó escalón útil en {u_col} para {tag}.')

    n0 = max(3, int(0.05 * len(y)))
    nf = max(3, int(0.10 * len(y)))
    y0 = float(np.mean(y[:n0]))
    yinf = float(np.mean(y[-nf:]))
    delta_y = yinf - y0
    K_static = delta_y / delta_u

    # --- Ziegler-Nichols por curva de reacción ---
    y_s = moving_average(y, window=9)
    dy = np.gradient(y_s, t)
    valid = np.arange(idx_step, len(t))
    if delta_y >= 0:
        i_inf = valid[int(np.argmax(dy[valid]))]
    else:
        i_inf = valid[int(np.argmin(dy[valid]))]

    m = float(dy[i_inf])
    yi = float(y_s[i_inf])
    ti = float(t[i_inf])

    if abs(m) < 1e-9:
        theta_zn = np.nan
        tau_zn = np.nan
    else:
        t_a = ti + (y0 - yi) / m
        t_b = ti + (yinf - yi) / m
        theta_zn = max(t_a - t0, 0.0)
        tau_zn = max(t_b - t_a, 1e-6)

    # --- Smith (28.3% y 63.2%) ---
    direction = 1 if delta_y >= 0 else -1
    y_28 = y0 + 0.283 * delta_y
    y_63 = y0 + 0.632 * delta_y
    t1 = first_crossing_time(t, y_s, y_28, idx_step, direction)
    t2 = first_crossing_time(t, y_s, y_63, idx_step, direction)

    if np.isnan(t1) or np.isnan(t2) or t2 <= t1:
        theta_sm = np.nan
        tau_sm = np.nan
    else:
        tau_sm = max(1.5 * (t2 - t1), 1e-6)
        theta_sm = max((1.5 * t1 - 0.5 * t2) - t0, 0.0)

    # --- Optimización numérica ---
    K_seed = K_static
    tau_seed = tau_sm if not np.isnan(tau_sm) else (tau_zn if not np.isnan(tau_zn) else max((t[-1] - t0) / 4, 1.0))
    theta_seed = theta_sm if not np.isnan(theta_sm) else (theta_zn if not np.isnan(theta_zn) else 0.0)

    t_fit = t[idx_step:]
    y_fit = y[idx_step:]

    def sse(params):
        K_, tau_, theta_ = params
        if tau_ <= 0 or theta_ < 0:
            return np.inf
        y_hat = fopdt_response(t_fit, K_, tau_, theta_, t0, delta_u, y0)
        e = y_fit - y_hat
        return float(np.mean(e**2))

    used_scipy = False
    try:
        from scipy.optimize import minimize
        bnds = [
            (K_seed - 3.0 * abs(K_seed) - 1.0, K_seed + 3.0 * abs(K_seed) + 1.0),
            (1e-6, max(10.0 * tau_seed, 1.0)),
            (0.0, max(t[-1] - t0, 1.0)),
        ]
        x0 = np.array([K_seed, max(tau_seed, 1e-3), max(theta_seed, 0.0)], dtype=float)
        sol = minimize(sse, x0=x0, method='L-BFGS-B', bounds=bnds)
        K_opt, tau_opt, theta_opt = sol.x
        used_scipy = bool(sol.success)
    except Exception:
        K_grid = np.linspace(K_seed - 1.5 * abs(K_seed) - 0.5, K_seed + 1.5 * abs(K_seed) + 0.5, 30)
        tau_grid = np.linspace(max(0.2 * tau_seed, 1e-3), max(3.0 * tau_seed, 2.0), 30)
        theta_grid = np.linspace(0.0, max(2.0 * theta_seed + 5.0, 20.0), 30)
        best = (K_seed, tau_seed, max(theta_seed, 0.0))
        best_cost = np.inf
        for K_ in K_grid:
            for tau_ in tau_grid:
                for theta_ in theta_grid:
                    c = sse((K_, tau_, theta_))
                    if c < best_cost:
                        best_cost = c
                        best = (K_, tau_, theta_)
        K_opt, tau_opt, theta_opt = best

    methods = {
        'Ziegler-Nichols': {'K': float(K_static), 'tau': float(tau_zn), 'theta': float(theta_zn)},
        'Smith': {'K': float(K_static), 'tau': float(tau_sm), 'theta': float(theta_sm)},
        'Optimización': {'K': float(K_opt), 'tau': float(tau_opt), 'theta': float(theta_opt)},
    }

    return {
        'tag': tag,
        't': t,
        'u': u,
        'y': y,
        't0': t0,
        'u0': u0,
        'uf': uf,
        'delta_u': delta_u,
        'y0': y0,
        'yinf': yinf,
        'delta_y': delta_y,
        'methods': methods,
        'used_scipy': used_scipy,
    }
```

```python
import numpy as np
# Identificación principal por canal dominante
id_q1 = estimate_fopdt_methods(df_q1, u_col='Cal1', y_col='Temp1', tag='Q1: Cal1 -> Temp1')
id_q2 = estimate_fopdt_methods(df_q2, u_col='Cal2', y_col='Temp2', tag='Q2: Cal2 -> Temp2')

rows = []
for ident in [id_q1, id_q2]:
    for method, pars in ident['methods'].items():
        rows.append({
            'Sistema': ident['tag'],
            'Método': method,
            'K (°C/%)': pars['K'],
            'tau (s)': pars['tau'],
            'theta (s)': pars['theta'],
        })

df_params = pd.DataFrame(rows)
display(df_params.round(4))

if id_q1['used_scipy'] or id_q2['used_scipy']:
    print('Optimización: se utilizó scipy.optimize.minimize (L-BFGS-B).')
else:
    print('Optimización: se utilizó búsqueda numérica en malla (sin scipy).')
```

```python
def plot_fopdt_fit(ident):
    t = ident['t']
    y = ident['y']
    t0 = ident['t0']
    y0 = ident['y0']
    du = ident['delta_u']

    plt.figure(figsize=(12, 5))
    plt.plot(t, y, 'k', linewidth=2, label='Medido')

    styles = {
        'Ziegler-Nichols': ('--', 'tab:blue'),
        'Smith': ('-.', 'tab:orange'),
        'Optimización': ('-', 'tab:green'),
    }

    for method, pars in ident['methods'].items():
        K = pars['K']
        tau = pars['tau']
        theta = pars['theta']
        if np.isnan(tau) or np.isnan(theta):
            continue
        y_hat = fopdt_response(t, K, tau, theta, t0, du, y0)
        ls, c = styles.get(method, ('--', None))
        plt.plot(t, y_hat, linestyle=ls, color=c, linewidth=2,
                 label=f"{method}: K={K:.3f}, τ={tau:.1f}s, θ={theta:.1f}s")

    plt.axvline(t0, color='gray', linestyle=':', label=f'Escalón de entrada t0={t0:.2f}s')
    plt.title(f"Comparación FOPDT - {ident['tag']}", fontweight='bold')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Temperatura (°C)')
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()


plot_fopdt_fit(id_q1)
plot_fopdt_fit(id_q2)
```
