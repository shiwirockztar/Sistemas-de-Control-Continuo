# Laboratorio 3 - Simulación de Sistemas de Control con Simulink

## Propósito
Este laboratorio busca conocer, con base en ejemplos, los procedimientos más comunes de la simulación de Sistemas de Control (SC) usando **Simulink (SL)**, la herramienta gráfica de MATLAB.

Además, se enfatiza en cómo conectar correctamente los espacios de trabajo de **MATLAB (ML)** y **Simulink (SL)** para analizar resultados de forma ordenada.

## Objetivos
- Comprender el flujo básico de modelado y simulación de un SC en Simulink.
- Aprender la comunicación entre MATLAB y Simulink.
- Identificar el propósito de comandos frecuentes de MATLAB para simulación y análisis: `lsim`, `sim`, `linmod`, `trim`, `simset`, `simget`.
- Reconocer la función de bloques esenciales de Simulink: `sign`, `in`, `out`, `to workspace`, `transfer fcn`.

## ¿Qué se entiende por simulación de un SC en Simulink?
Simular un sistema de control consiste en representar matemáticamente la planta, el controlador y las señales de referencia/perturbación para observar su respuesta temporal y evaluar desempeño.

Con Simulink se construye un diagrama de bloques que implementa ecuaciones diferenciales y relaciones de transferencia, permitiendo:
- validar conceptos teóricos,
- comparar configuraciones de control,
- estimar estabilidad y desempeño antes de pruebas reales.

## Conexión entre los espacios de trabajo de MATLAB (ML) y Simulink (SL)
La interacción ML-SL permite preparar parámetros desde scripts y recuperar resultados de simulación para análisis posterior.

Flujo típico:
1. Definir parámetros y señales en MATLAB (`Ts`, constantes, vectores de tiempo, entradas).
2. Ejecutar el modelo de Simulink con `sim`.
3. Exportar señales desde bloques como `To Workspace`.
4. Analizar resultados en MATLAB (gráficas, métricas, comparación con teoría).

Buenas prácticas:
- Nombrar de forma clara variables y señales exportadas.
- Mantener en un script `.m` la parametrización para repetir experimentos.
- Verificar unidades y tiempos de muestreo antes de cada simulación.

## Descripción somera de comandos de MATLAB

| Comando | Función principal | Uso típico en el laboratorio |
|---|---|---|
| `lsim` | Simula la respuesta temporal de un modelo lineal ante una entrada arbitraria. | Evaluar respuesta de una función de transferencia o modelo en espacio de estados con señales definidas por el usuario. |
| `sim` | Ejecuta un modelo de Simulink desde MATLAB. | Correr experimentos automáticos cambiando parámetros desde scripts. |
| `linmod` | Obtiene una aproximación lineal (modelo linealizado) de un modelo de Simulink alrededor de un punto de operación. | Comparar comportamiento no lineal vs lineal para análisis de control clásico. |
| `trim` | Calcula un punto de equilibrio (estado y entradas) que satisface condiciones de operación estacionaria. | Hallar condiciones de operación antes de linealizar o diseñar controladores. |
| `simset` | Configura opciones de simulación (método numérico, tolerancias, tiempos, etc.). | Ajustar precisión y condiciones de integración para resultados confiables. |
| `simget` | Consulta opciones de simulación actualmente definidas. | Verificar o inspeccionar la configuración activa antes de ejecutar pruebas. |

## Descripción somera de bloques de Simulink

| Bloque | Función | Observaciones |
|---|---|---|
| `sign` | Entrega el signo de la señal de entrada: $-1$, $0$ o $1$. | Útil en lógica de control por regiones o funciones no lineales simples. |
| `in` (`Inport`) | Define una entrada al subsistema/modelo. | Permite conectar señales externas de manera organizada. |
| `out` (`Outport`) | Define una salida del subsistema/modelo. | Facilita modularidad y lectura del diagrama. |
| `to workspace` (`To Workspace`) | Exporta señales desde Simulink hacia MATLAB. | Fundamental para graficar y procesar resultados en ML. |
| `transfer fcn` | Implementa una función de transferencia continua del tipo $\frac{N(s)}{D(s)}$. | Se usa para modelar plantas o controladores lineales en dominio de Laplace. |

## Mini guía de procedimiento (resumen)
1. Definir en MATLAB la planta y condiciones iniciales.
2. Construir en Simulink el diagrama con `Inport`, `Transfer Fcn`, `Sign`, `Outport` y `To Workspace` según corresponda.
3. Configurar opciones de simulación (`simset`) y verificarlas (`simget`).
4. Ejecutar simulación (`sim`) y, cuando aplique, contrastar con respuesta lineal (`lsim`).
5. Si se requiere análisis local, obtener punto de equilibrio (`trim`) y linealización (`linmod`).
6. Comparar resultados, extraer conclusiones y documentar limitaciones del modelo.

## Actividad 4.3: Scope, Logging y visualizacion de senales
1. Haga doble click sobre el **Floating Scope** y recorra los iconos del menu para identificar su funcion.
2. En la ventana del Scope vaya a **View >> Configuration Properties** y revise sus pestanas.
3. En **Logging**, habilite el envio de datos al Espacio de Trabajo de MATLAB (ETML/workspace).

Formatos de guardado frecuentes:
- `Structure with time`
- `Structure`
- `Array`

Para esta practica se recomienda usar **Scope normal (no flotante)** y agregar un bloque **Mux** de 3 entradas para observar simultaneamente:
- entrada del sistema,
- salida del sistema,
- accion de control.

Conexion sugerida:
- entradas del Mux: entrada, salida, control.
- salida del Mux: Scope.

Ayuda adicional desde la VCML:
- `doc scope`
- `doc <nombre_bloque>` para consultar bloques/iconos de Simulink.

## Estudio de bloques: In1, Out1 y To Workspace
Actividad sugerida para preinforme:
1. Abra la libreria de Simulink.
2. Haga doble click en `In1 (Inport)`, `Out1 (Outport)` y `To Workspace`.
3. Revise parametros y opciones de cada bloque.

Funcion de cada bloque:
- **In1 (Inport):** recibe una senal de entrada hacia un subsistema o modelo.
- **Out1 (Outport):** entrega una senal de salida desde un subsistema o modelo.
- **To Workspace:** exporta senales al ETML para analisis posterior.

### Diferencia entre Out y To Workspace
- **Out/Outport** define la **interfaz de salida** del diagrama (conexion y arquitectura del modelo).
- **To Workspace** no define interfaz; solo **almacena datos de simulacion** en MATLAB.

Conclusiones:
- Use Outport cuando necesite exponer la senal como puerto del modelo.
- Use To Workspace cuando necesite graficar, comparar o postprocesar resultados en MATLAB.
- En un mismo modelo pueden coexistir ambas opciones.

## Actividad 4.5: Parametrizacion, simulacion y analisis en ETML
Para correr correctamente el modelo, primero parametrice el entorno de simulacion en:
**Modeling >> Model Settings >> Model Settings**

Opciones clave a estudiar:
1. **Solver**
- Metodo de solucion.
- Paso fijo o variable.
- Tiempo de simulacion y parametros numericos.

2. **Data Import/Export**
- Definir si se reciben datos desde ETML (por ejemplo, con `In1`).
- Definir que resultados se exportan al ETML (por ejemplo, con `Out1`).
- Para esta practica, guardar `time`, `states` y `output` en formato **Array**.

3. **Diagnostics**
- Definir condiciones de chequeo, alarmas y paro de simulacion.

### Verificacion previa del workspace
Antes de simular, en la VCML ejecute `whos` para verificar variables existentes.
- Si hay variables anteriores, borrelas selectivamente.
- Evite `clear all` para no perder configuraciones utiles de trabajo.

### Despues de simular
1. Active el Scope y observe las senales.
2. Identifique variables generadas en ETML.
3. Relacione cada variable con la configuracion elegida en Model Settings y/o bloques del diagrama.
4. Verifique si aparecen variables no esperadas.

### Graficar datos cuando estan en formato array
Si una variable viene como matriz:
- primera columna: tiempo,
- siguientes columnas: senales.

Ejemplo conceptual:
- `t = output(:,1)`
- `y = output(:,2)`
- `plot(t,y)`

La extraccion se hace por indexacion de columnas de matriz, por ejemplo `M(:,k)`.
