# Parcial 1 - Funcion de Transferencia

Este notebook muestra el proceso completo para obtener la funcion de transferencia del sistema resorte-masa-polea.

## 1. Ecuacion de movimiento

Partimos de la ecuacion dinamica obtenida por energia:

$$\left(m+\frac{3}{8}M\right)\ddot{x}+\frac{1}{4}kx=0$$

Para definir una funcion de transferencia entrada/salida, se considera una fuerza externa $F(t)$:

$$\left(m+\frac{3}{8}M\right)\ddot{x}+\frac{1}{4}kx=F(t)$$

## 1.1 Derivacion detallada de la ecuacion de movimiento

Partimos de las relaciones cinematicas del sistema:

$$x=2y, \quad R\theta = x-y=y, \quad J=\frac{1}{2}MR^2$$

De aqui se obtiene:

$$y=\frac{x}{2}, \quad \dot{y}=\frac{\dot{x}}{2}, \quad \dot{\theta}=\frac{\dot{y}}{R}=\frac{\dot{x}}{2R}$$

### Paso 1: Energia cinetica

$$T=\frac{1}{2}m\dot{x}^2+\frac{1}{2}M\dot{y}^2+\frac{1}{2}J\dot{\theta}^2$$

Sustituyendo $\dot{y}=\dot{x}/2$, $J=\frac{1}{2}MR^2$ y $\dot{\theta}=\dot{x}/(2R)$:

$$
T=\frac{1}{2}m\dot{x}^2+\frac{1}{2}M\left(\frac{\dot{x}}{2}\right)^2+\frac{1}{2}\left(\frac{1}{2}MR^2\right)\left(\frac{\dot{x}}{2R}\right)^2
$$

$$
T=\frac{1}{2}m\dot{x}^2+\frac{1}{8}M\dot{x}^2+\frac{1}{16}M\dot{x}^2
=\frac{1}{2}m\dot{x}^2+\frac{3}{16}M\dot{x}^2
$$

### Paso 2: Energia potencial

Desde el equilibrio:

$$U_0=\frac{1}{2}ky_\delta^2+Mg(l-y_0)+mg(l-x_0)$$

Para desplazamientos $x$ y $y$:

$$U=\frac{1}{2}k(y_\delta+y)^2+Mg(l-y_0-y)+mg(l-x_0-x)$$

Al expandir:

$$U=U_0+\frac{1}{2}ky^2+ky_\delta y-Mgy-mgx$$

Condicion estatica del resorte:

$$ky_\delta=Mg+2mg$$

Multiplicando por $y$ y usando $x=2y$:

$$ky_\delta y=Mgy+2mgy=Mgy+mgx$$

Entonces se cancelan los terminos gravitacionales lineales y queda:

$$U=U_0+\frac{1}{2}ky^2=U_0+\frac{1}{8}kx^2$$

### Paso 3: Conservacion de energia

$$T+U=\text{constante}$$

$$\frac{1}{2}m\dot{x}^2+\frac{3}{16}M\dot{x}^2+U_0+\frac{1}{8}kx^2=\text{constante}$$

Derivando respecto al tiempo:

$$\left(m+\frac{3}{8}M\right)\dot{x}\ddot{x}+\frac{1}{4}kx\dot{x}=0$$

Factorizando $\dot{x}$:

$$\left[\left(m+\frac{3}{8}M\right)\ddot{x}+\frac{1}{4}kx\right]\dot{x}=0$$

Para movimiento no trivial ($\dot{x}\neq0$):

$$\left(m+\frac{3}{8}M\right)\ddot{x}+\frac{1}{4}kx=0$$

Forma normalizada:

$$\ddot{x}+\frac{2k}{8m+3M}x=0$$

Por comparacion con $\ddot{x}+\omega_n^2x=0$:

$$\omega_n=\sqrt{\frac{2k}{8m+3M}}$$

Esta es exactamente la frecuencia natural pedida por el enunciado usando conservacion de energia.

```python
import sympy as sp

s, m, M, k = sp.symbols('s m M k', positive=True, real=True)
X, F = sp.symbols('X F')

m_eq = m + sp.Rational(3, 8) * M
print('m_eq =', m_eq)
```

## 2. Transformada de Laplace (condiciones iniciales cero)

Aplicando Laplace a la ecuacion con entrada:

$$\left[\left(m+\frac{3}{8}M\right)s^2+\frac{1}{4}k\right]X(s)=F(s)$$

```python
G = sp.simplify(1 / (m_eq * s**2 + sp.Rational(1, 4) * k))
G_eq = sp.simplify(8 / ((8*m + 3*M) * s**2 + 2*k))

print('G(s) forma directa:')
sp.pprint(G)

print('\nG(s) forma equivalente:')
sp.pprint(G_eq)

print('\nDiferencia simbolica G - G_eq =', sp.simplify(G - G_eq))
```

## 3. Funcion de transferencia

La funcion de transferencia desplazamiento-fuerza es:

$$G(s)=\frac{X(s)}{F(s)}=\frac{1}{\left(m+\frac{3}{8}M\right)s^2+\frac{1}{4}k}$$

Forma equivalente:

$$G(s)=\frac{8}{(8m+3M)s^2+2k}$$

```python
omega_n = sp.sqrt(2*k / (8*m + 3*M))

print('Frecuencia natural:')
sp.pprint(omega_n)

G_std = sp.simplify((1/m_eq) / (s**2 + omega_n**2))
print('\nForma estandar de segundo orden:')
sp.pprint(G_std)

print('\nVerificacion G - G_std =', sp.simplify(G - G_std))
```

## 4. Resultado final

$$\omega_n=\sqrt{\frac{2k}{8m+3M}}$$

$$G(s)=\frac{X(s)}{F(s)}=\frac{1}{\left(m+\frac{3}{8}M\right)s^2+\frac{1}{4}k}=\frac{8}{(8m+3M)s^2+2k}$$
