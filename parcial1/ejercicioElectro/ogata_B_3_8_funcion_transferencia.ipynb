# Ogata B-3-8: Obtencion de $E_o(s)/E_i(s)$

En este notebook se desarrolla la funcion de transferencia del circuito usando las ecuaciones de malla dadas.

## 1) Ecuaciones en tiempo

Ecuacion 1 (malla izquierda):
$$L(\ddot q_1-\ddot q_2)+R_1\dot q_1+R_2(\dot q_1-\dot q_2)=e_i$$

Ecuacion 2 (malla derecha):
$$L(\ddot q_2-\ddot q_1)+R_2(\dot q_2-\dot q_1)+\frac{q_2}{C}=0$$

Ecuacion de salida:
$$e_o=\frac{q_2}{C}$$

## 2) Transformada de Laplace (condiciones iniciales cero)

Definiendo $Q_1(s),Q_2(s),E_i(s),E_o(s)$:
$$
(Ls^2+(R_1+R_2)s)Q_1-(Ls^2+R_2s)Q_2=E_i
$$
$$
-(Ls^2+R_2s)Q_1+\left(Ls^2+R_2s+\frac{1}{C}\right)Q_2=0
$$
$$E_o=\frac{Q_2}{C}$$

```python
import sympy as sp

s, L, R1, R2, C = sp.symbols('s L R1 R2 C', positive=True)
Q1, Q2, Ei = sp.symbols('Q1 Q2 Ei')

eq1 = sp.Eq((L*s**2 + (R1+R2)*s)*Q1 - (L*s**2 + R2*s)*Q2, Ei)
eq2 = sp.Eq(-(L*s**2 + R2*s)*Q1 + (L*s**2 + R2*s + 1/C)*Q2, 0)

sol = sp.solve([eq1, eq2], (Q1, Q2), dict=True)[0]
Q2_over_Ei = sp.simplify(sp.factor(sol[Q2] / Ei))
G = sp.simplify(sp.factor((Q2_over_Ei) / C))  # G(s)=Eo/Ei

sp.pretty(G)
```

## 3) Funcion de transferencia resultante

Una forma compacta es:
$$
\boxed{\frac{E_o(s)}{E_i(s)}=\frac{Ls+R_2}{C R_1 L s^2 + (L + C R_1 R_2)s + (R_1+R_2)}}
$$

Forma equivalente (sin cancelar $s$ en una etapa intermedia):
$$
\frac{E_o(s)}{E_i(s)}=\frac{s(Ls+R_2)}{s\left[C R_1 L s^2 + (L + C R_1 R_2)s + (R_1+R_2)\right]}
$$

```python
G_objetivo = (L*s + R2) / (C*R1*L*s**2 + (L + C*R1*R2)*s + (R1 + R2))
sp.simplify(G - G_objetivo)
```
